```python
# =========================
# IMPORT
# =========================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import joblib
```

```python
# =========================
# LOAD DATA
# =========================
df = pd.read_csv("rotten_tomatoes_movies.csv")
```

```python
# =========================
# SELECT FEATURES
# =========================
df = df[['tomatometer', 'audience_score', 'year']].dropna()
```

```python
# =========================
# DEFINE X, y
# =========================
X = df[['tomatometer', 'year']]
```

```python
# target: หนังดีหรือไม่
y = (df['audience_score'] > 60).astype(int)
```

```python
# =========================
# SPLIT
# =========================
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# =========================
# SCALE
# =========================
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# =========================
# MODEL
# =========================
model = Sequential([
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
# =========================
# COMPILE
# =========================
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```

```python
# =========================
# TRAIN
# =========================
model.fit(X_train, y_train, epochs=15, batch_size=32)
```

```python
# save architecture
model_json = model.to_json()
with open("nn_model.json", "w") as json_file:
    json_file.write(model_json)

# save weights (ชื่อไฟล์ต้องแบบนี้เท่านั้น)
model.save_weights("nn_model.weights.h5")

print("Saved JSON + Weights!")
```

```python
from sklearn.neural_network import MLPClassifier
import joblib

model = MLPClassifier(hidden_layer_sizes=(128,64,32), max_iter=500)
model.fit(X_train, y_train)

joblib.dump(model, "nn_model.pkl")
joblib.dump(scaler, "nn_scaler.pkl")
```
